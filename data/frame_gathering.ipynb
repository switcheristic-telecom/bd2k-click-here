```python
# read clickheres.csv

import csv

ch_banners = []

with open('clickheres.csv', 'r') as f:
    # read all lines into a list of dictionaries
    reader = csv.DictReader(f)
    for row in reader:
        ch_banners.append(row)

# print the first 5 rows
for i in range(5):
    print(ch_banners[i])
```

```python
ch_banners_en = [banner for banner in ch_banners if banner['language'] == 'en']
ch_banners_es = [banner for banner in ch_banners if banner['language'] == 'es']
ch_banners_zh = [banner for banner in ch_banners if banner['language'] == 'zh']
```

```python
for banner in ch_banners_en:
    print(banner)
```

```python
import requests
url_template = "https://banner-depot-2000.s3.us-west-1.amazonaws.com/banner-image/{md5}/{md5}{frame}.gif"
# download all ch_banners_en images to directory 'en'
import os
os.makedirs('en', exist_ok=True)
for banner in ch_banners_en:
    _frame = '-' + banner['frame']
    print(f"frame: {_frame}")
    if _frame == '--1':
        _frame = ''
    url = url_template.format(md5=banner['md5'], frame=_frame)
    print(f"url: {url}")
    response = requests.get(url)
    print(f"downloading {banner['md5']}-{banner['frame']}.gif")
    with open(f"en/{banner['md5']}{_frame}.gif", 'wb') as f:
        f.write(response.content)
        print(f"saved {banner['md5']}-frame.gif")
```
