```python
# this script takes an HTML string as input, finds the first <meta> tag in the html, and inserts a new <script> tag pointing to a JavaScript file located at "../additional-script.js", and returns the modified HTML string

import bs4

def insert_script_tag(html):
    soup = bs4.BeautifulSoup(html, 'html.parser')
    meta_tag = soup.find('meta')
    script_tag = soup.new_tag('script', src='../additional-script.js')
    meta_tag.insert_after(script_tag)
    return str(soup)
```

```python
HTML_DIRECTORY = "html_es/"

# gather all HTML files in the directory specified
import os
html_files = [f for f in os.listdir(HTML_DIRECTORY) if f.endswith('.html')]

for html_file in html_files:
    # read the HTML file
    with open(HTML_DIRECTORY + html_file, 'r') as f:
        html_string = f.read()
    # insert the script tag
    modified_html_string = insert_script_tag(html_string)
    # write the modified HTML file
    with open(HTML_DIRECTORY + html_file, 'w') as f:
        f.write(modified_html_string)
    print(f"Processed {html_file}")
```
