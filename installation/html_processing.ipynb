```python
# this script takes an HTML file and an MD5 hash as input, 
# finds the <img> tag that contains the base64 of a GIF image that has the MD5 hash, 
# gives the <img> tag a new class named "bd-2k-banner", 
# then, it adds a link to a CSS file that contains the CSS rule for the new class, located at "../additional-style.css"
# and finally, writes the modified HTML file to the output file

import sys
import re
import base64
import hashlib
import bs4

def html_processor(html_string, md5_string): 
    # parse the HTML string
    soup = bs4.BeautifulSoup(html_string, 'html.parser')
    # find the <img> tag that contains the base64 of a GIF image that has the MD5 hash
    for img_tag in soup.find_all('img'):
        if img_tag.has_attr('src') and img_tag['src'].startswith('data:image/gif;base64,'):
            # decode the base64 string
            img_base64 = img_tag['src'][len('data:image/gif;base64,'):]
            img_bytes = base64.b64decode(img_base64)
            # calculate the MD5 hash of the image
            img_md5 = hashlib.md5(img_bytes).hexdigest()
            # if the MD5 hash matches the input MD5 hash
            if img_md5 == md5_string:
                # add a new class to the <img> tag, named "bd-2k-banner"
                img_tag['class'] = img_tag.get('class', []) + ['bd-2k-banner']
    # add a link to a CSS file that contains the CSS rule for the new class, located at
    # "../additional-style.css"
    css_link_tag = soup.new_tag('link')
    css_link_tag['rel'] = 'stylesheet'
    css_link_tag['type'] = 'text/css'
    css_link_tag['href'] = '../additional-style.css'
    # soup.head.append(css_link_tag)
    # these HTML files don't have <head> or <body> tags, so we append the after the first <meta> tag
    # find the fist <meta> tag
    first_meta_tag = soup.find('meta')
    # insert the CSS link tag after the first <meta> tag
    first_meta_tag.insert_after(css_link_tag)
    # output the modified HTML source as string
    return str(soup)
```

```python
HTML_DIRECTORY = "html_ja/"

# gather all HTML files in the directory specified
import os
html_files = [f for f in os.listdir(HTML_DIRECTORY) if f.endswith('.html')]

for html_file in html_files:
    # read the HTML file
    with open(HTML_DIRECTORY + html_file, 'r') as f:
        html_string = f.read()
    # read the MD5 hash
    md5_string = html_file.split('.')[0]
    # process the HTML file
    modified_html_string = html_processor(html_string, md5_string)
    # write the modified HTML file
    with open(HTML_DIRECTORY + html_file, 'w') as f:
        f.write(modified_html_string)
    print(f"Processed {html_file}")
```
