```python
# For zh and en languages, the following code is to identify and download MHTML files that contain the banner ad images
# First, identify language to process
# 

LANGUAGE_TO_PROCESS = "en"  
```

```python
# gather a list of banner MD5s from the directory of GIF files for the language
import os
import shutil

# get all .gif files in the language directory
banner_md5s = []
for root, dirs, files in os.walk(LANGUAGE_TO_PROCESS):
    for file in files:
        if file.endswith(".gif"):
            # get the part of the file name before the first -
            banner_md5 = file.split("-")[0]
            banner_md5s.append(banner_md5)

print(banner_md5s)
```

```python
# load banners_output_20230930.json
import json
with open('banners_output_20230930.json') as f:
    banners_output = json.load(f)

# for each banner image md5 in banner_md5s, get the list of urls from banners_output.appearances
banner_urls = []
for banner_md5 in banner_md5s:
    # find the object in banners_output that has the same md5
    for banner in banners_output:
        if banner['md5'] == banner_md5:
            banner_urls.append({
                'md5': banner_md5,  
                'url': banner['appearances'][0]['url'], 
                'timestamp': banner['appearances'][0]['timestamp']
                })
            break

# output banner_urls to a JSON
with open('mhtml_results_{}.json'.format(LANGUAGE_TO_PROCESS), 'w') as f:
    json.dump(banner_urls, f, indent=4)
```
